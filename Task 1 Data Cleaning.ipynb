# DATA CLEANING

****Importing dataset****

```python
import pandas as pd
df=pd.read_csv('marketing_campaign.csv',sep='\t', encoding='utf-8')
```

****Basic details****

```python
df.head()
```

****Handling missing values****

```python
print(df.isnull())
```

****Dropping duplicate values****

```python
print(df.drop_duplicates())
```

****Standardizing text values****

```python
df['Education']=df['Education'].str.lower()
df['Marital_Status']=df['Marital_Status'].str.lower()
```

****Converting date formats to a consistent type

```python
from datetime import datetime
current_year = datetime.now().year
df['age'] = current_year - df['year_birth']
```

```python
df['age'] = df['age'].astype(int)
df['dt_customer'] = pd.to_datetime(df['dt_customer'], format='%d-%m-%Y', errors='coerce')
```

****Renaming column headers to be clean and uniform****

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

****Fixing data types****

```python
df['year_birth'] = pd.to_numeric(df['year_birth'], errors='coerce')
df['income'] = pd.to_numeric(df['income'], errors='coerce')
```

```python
print(df.dtypes)
```

```python
df.to_csv("cleaned_marketing_campaign.csv", index=False)
```
